```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
dx=0.01
x=np.arange(0,1,dx)
n=len(x)
def model1(u):
    dydx=u
    return dydx
def model2(y):
    dudx=y
    return dudx
```

```python
y=np.zeros(n)
u=np.zeros(n)
y[0]=2
k1=np.zeros(n)
k2=np.zeros(n)
k3=np.zeros(n)
k4=np.zeros(n)
t1=np.zeros(n)
t2=np.zeros(n)
t3=np.zeros(n)
t4=np.zeros(n)
u[0]=-1
for i in range(0,n-1):
    
    k1[i]=model1(u[i])
    k2[i]=model1(u[i]+0.5*dx*k1[i])
    k3[i]=model1(u[i]+0.5*dx*k2[i])
    k4[i]=model1(u[i]+dx*k3[i])
    y[i+1]=y[i]+dx*((k1[i]+2*k2[i]+2*k3[i]+k4[i])/6)
    t1[i]=model2(y[i+1])
    t2[i]=model2(y[i+1]+0.5*dx*t1[i])
    t3[i]=model2(y[i+1]+0.5*dx*t2[i])
    t4[i]=model2(y[i+1]+dx*t3[i])
    u[i+1]=u[i]+dx*((t1[i]+2*t2[i]+2*t3[i]+t4[i])/6)
                   
                   
y
```

```python
plt.plot(x,y)
plt.grid()
```

```python
from scipy.integrate import odeint
t=np.linspace(0,1,101)
def derivative(z,t):
    p=z[0]
    v=z[1]
    dpdt=v
    dvdt=p
    return dpdt,dvdt
z0=[2,-1]
sol=odeint(derivative,z0,t)

p=sol[:,0]
p
plt.plot(t,p)
plt.plot(x,y)
plt.grid()
```
