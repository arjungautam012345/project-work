```python
from scipy.integrate import odeint
from matplotlib import pyplot as plt
import numpy as np
```

```python
dx=0.01
x=np.arange(0,1,dx)
n=len(x)
n
```

```python
def model1(y):
    dudx=y
    return dudx
```

```python
def model2(u):
    dydx=u
    return dydx
    
```

```python
u=np.zeros(n)
y=np.zeros(n)
y[0]=2
u[0]=-1
for i in range(0,n-1):
    y[i+1]=y[i]+dx*model2(u[i])
    u[i+1]=u[i]+dx*model1(y[i+1])
    

y
```

```python

plt.plot(x,y)
plt.grid()
```

```python
t=np.linspace(0,1,100)
def derivative(z,t):
    p=z[0]
    v=z[1]
    dpdt=v
    dvdt=p
    return dpdt,dvdt
z0=[2,-1]
sol=odeint(derivative,z0,t)

p=sol[:,0]
p
plt.plot(t,p)
plt.plot(x,y)
plt.grid()
    
```
