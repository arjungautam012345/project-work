```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dx=0.01
x=np.arange(-1,1+dx,dx)
N=x.size
```

```python
k=1
def V(x):
    return 0.5*k*x**2

    
```

```python

m=1
hbar=1
def k2(e,x):
    k=(2*m/hbar**2)*(e-V(x))
    return k
    
 
    
    
```

```python
psi=np.zeros(N)
b=(dx**2)/12
def wavefunction(e):
    for i in range(1,N-1):
        psi[0]=0
        psi[1]=1e-5
        psi[i+1]=(2*psi[i]*(1-5*b*k2(e,x[i]))-(1+b*k2(e,x[i-1]))*psi[i-1])/(1+b*k2(e,x[i+1]))
    return psi
    
```

```python
a1=wavefunction(40)
plt.plot(x,a1)
plt.plot(x,V(x)*0.001)
plt.grid()
```

```python
def countnodes(e):
    count=0
    for i in range(1,N-1):
        if np.sign(wavefunction(e)[i])!=np.sign(wavefunction(e)[i+1]):
            count=count+1
    return count

 
```
